```python
import random
import matplotlib.pyplot as plt

# Defining Mesh Nodes with their user capacities
mesh_nodes = {
    "Node_A": {"capacity": 50, "current_users": random.randint(20, 50)},
    "Node_B": {"capacity": 50, "current_users": random.randint(10, 50)},
    "Node_C": {"capacity": 50, "current_users": random.randint(5, 50)}
}

def assign_user_to_node():
    for node, data in mesh_nodes.items():
        if data["current_users"] < data["capacity"]:
            data["current_users"] += 1
            return node
    return "No Available Node"

# Simulating 30 new users trying to connect
new_connections = 30
assignments = {"Node_A": 0, "Node_B": 0, "Node_C": 0, "Dropped": 0}

for _ in range(new_connections):
    assigned_node = assign_user_to_node()
    if assigned_node != "No Available Node":
        assignments[assigned_node] += 1
    else:
        assignments["Dropped"] += 1  # Users dropped due to overload

# Visualizing Load Distribution
plt.bar(assignments.keys(), assignments.values(), color=['blue', 'green', 'red', 'black'])
plt.xlabel("Mesh Nodes")
plt.ylabel("New Connections Assigned")
plt.title("User Load Balancing in ECCRS")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Sample training data (real vs fake alerts)
alerts = [
    "Bomb blast in city center, need immediate help!",  # Real
    "Flood warning issued for entire region, evacuate now!",  # Real
    "Aliens landed in the capital, total chaos!",  # Fake
    "Zombies attacking citizens, urgent military action needed!",  # Fake
]

labels = [1, 1, 0, 0]  # 1 = Real, 0 = Fake

# Training Naive Bayes classifier
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(alerts)
model = MultinomialNB()
model.fit(X_train, labels)

# Test new alert
new_alert = ["Heavy rains causing floods in district, immediate rescue needed!"]
X_test = vectorizer.transform(new_alert)
prediction = model.predict(X_test)

print("🔹 Alert Classification:", "REAL" if prediction[0] == 1 else "FAKE")
```

```python
restricted_sites = ["fake-news.com", "socialmedia.com", "banned-forum.net"]
allowed_sites = ["govt-alerts.in", "disaster-relief.org", "emergency.gov"]

def check_access(site):
    if site in restricted_sites:
        return "❌ Access Denied"
    elif site in allowed_sites:
        return "✅ Access Granted"
    else:
        return "⚠️ Unknown Site - Under Review"

# Simulating user request
site_request = "govt-alerts.in"
print(f"🔍 Checking access for {site_request}: {check_access(site_request)}")
```
